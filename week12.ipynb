실습 1 : 서울의 일교차

```python
import csv
f= open('seoul_temp.csv','r',encoding = 'cp949')
data = csv.reader(f,delimiter=',')
header = next(data)
max_temp = -999
max_temp_date = 0
min_temp = 10000
mintemp_date = 0
max_dif = -999    
max_dif_date = 0

for row in data:
    if row[-1]=='':
        row[-1]= -999
    row[-1] = float(row[-1])
    if(max_temp < row[-1]):
        max_temp_date = row[0]
        max_temp = row[-1]
    if row[-2] == '':
        row[-2] = 0 
    row[-2] = float(row[-2])
    if(min_temp > row[-2]):
        min_temp_date= row[0]
        min_temp = row[-2]
    dif = row[-1] - row[-2]
    if(max_dif < dif):
        max_dif_date = row[0]
        max_dif = dif
print("일교차가 가장 큰 날: {0:s}".format(max_dif_date))
print("최저 기온이 가장 낮은 날 : {0:s}".format(min_temp_date))
print("최고 기온이 가장 높은 날 : {0:s}".format(max_temp_date))  
f.close()
```

실습 2 : 지하철 이용객

```python
import csv
f= open('2021년 12월  교통카드 통계자료.csv','r',encoding = 'utf-8')
data = csv.reader(f,delimiter=',')
header = next(data)
linearr = []
data2 = []
eachline = []
for line in data:#어떤 라인 있는지 받아오기
    if(line[1] not in linearr):
        linearr.append(line[1])
    data2.append(line) # 데이터에 다시 넣기
    
for subway in linearr:#각각의 라인에 대해서 data for_loop
    print("=="+subway+"==")
    Arr = []
    Idx = ""
    for col in data2:
        if (subway == col[1]):
            passenger =int(col[4].replace(",",""))+int(col[5].replace(",",""))
            Arr.append([col[3],passenger])
    Arr.sort(key = lambda x:x[1])
    print(" 이용객이 가장 많은 역 순위")
    print(Arr[-1])
    print(Arr[-2])
    print(Arr[-3])
    print(" 이용객이 가장 적은 역 순위")
    print(Arr[0])
    print(Arr[1])
    print(Arr[2])    
f.close()
```
